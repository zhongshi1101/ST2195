```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
flight_2008 = pd.read_csv('./dataverse_files/2008.csv')
flight_2007 = pd.read_csv('./dataverse_files/2007.csv')
flight_2006 = pd.read_csv('./dataverse_files/2006.csv')

flight = pd.concat([flight_2006, flight_2007, flight_2008])
flight.head()
```

```python
len(flight)
flight_sub = flight
```

```python
flight_sub_dest_grouped = flight_sub.groupby(['Year', 'Month', 'DayofMonth', 'Dest'], as_index=False).agg({'DepDelay': 'mean'})
flight_sub_orig_grouped = flight_sub.groupby(['Year', 'Month', 'DayofMonth', 'Origin'], as_index=False).agg({'DepDelay': 'mean'})
```

```python
flight_sub_joined = pd.merge(flight_sub_dest_grouped, flight_sub_orig_grouped, how='inner', left_on=['Year', 'Month', 'DayofMonth', 'Dest'], right_on=['Year', 'Month', 'DayofMonth', 'Origin'])
flight_sub_joined.head()
```

- merge by same year, same month, same day of mont, A.arrTime < B.depTime, A.dest = B.origin
- compute delay ratio
